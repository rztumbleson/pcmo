```python
from scipy.ndimage import gaussian_filter1d
```

```python
plt.plot(np.arange(len(df220['sigmax']))*1.625, df220['sigmax']/df220['sigmay'], 'o', alpha=0.05, markersize=3, color='C0')
plt.plot(np.arange(len(df220['sigmax']))*1.625, gaussian_filter1d(df220['sigmax']/df220['sigmay'], sigma=21), color='C0', label='220 K')
plt.plot(np.arange(len(df220['sigmax']))*1.625, df225['sigmax']/df225['sigmay'], 'o', alpha=0.05, markersize=3, color='C1')
plt.plot(np.arange(len(df220['sigmax']))*1.625, gaussian_filter1d(df225['sigmax']/df225['sigmay'], sigma=21), color='C1', label='225 K')
plt.plot(np.arange(len(df220['sigmax']))*1.625, df230['sigmax']/df230['sigmay'], 'o', alpha=0.05, markersize=3, color='C2')
plt.plot(np.arange(len(df220['sigmax']))*1.625, gaussian_filter1d(df230['sigmax']/df230['sigmay'], sigma=21), color='C2', label='230 K')
plt.plot(np.arange(len(df230_2['sigmax']))*1.625, df230_2['sigmax']/df230_2['sigmay'], 'o', alpha=0.05, markersize=3, color='C3')
plt.plot(np.arange(len(df230_2['sigmax']))*1.625, gaussian_filter1d(df230_2['sigmax']/df230_2['sigmay'], sigma=21), color='C3', label='230 K - 2 hours')
plt.legend(fontsize=14)
plt.xlabel('Time (s)')
plt.ylabel(r'$\sigma_x / \sigma_y$')
plt.show()
```
